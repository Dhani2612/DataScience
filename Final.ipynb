```python
bad_loan = [
    'Charged Off',
    'Late (31-120 days)',
    'Late (16-30 days)',
    'Default',
    'Does not meet the credit policy. Status:Charged Off'
]

df2= df.copy()

df2['good/bad'] = np.where(df['loan_status'].isin(bad_loan), 0, 1)
df2['good/bad'].value_counts(normalize=True)*100
```

```python
import plotly.graph_objects as go

# Menghitung jumlah setiap kategori
df_type = df.groupby('loan_status')['id'].count().reset_index()
df_type = df_type.rename(columns={"id": "count_showids"})

# Menghitung total dan persentase
total = df_type['count_showids'].sum()
df_type['percentage'] = (df_type['count_showids'] / total * 100).round(2)

# Membuat teks yang berisi jumlah dan persentase
text_labels = [f"{count} ({perc}%)" for count, perc in zip(df_type['count_showids'], df_type['percentage'])]

layout = go.Layout(title='Loan Status')

fig = go.Figure(layout=layout)
fig.add_trace(
    go.Bar(
        x=['0 (Bad Loan)', '1 (Good Loan)'],
        y=df_type['count_showids'],
        marker=dict(color=["#0c2054", "#f3ac3d"], line=dict(width=2, color='black')),
        text=text_labels,  # Menampilkan jumlah dan persentase
        textposition='auto',
        opacity=0.8
    ))

fig.update_layout(
    title='Loan Status',
    title_font_family="Verdana",
    title_font_color="#222222",
    title_font_size=28,
    title_x=0.5,
    width=800,
    height=500,
    plot_bgcolor="whitesmoke",
    xaxis=dict(title="Loan Status", tickfont=dict(size=14, family="Arial")),
    yaxis=dict(title="Count", tickfont=dict(size=14, family="Arial")),
    legend=dict(
        x=0.8,
        y=1,
        title_font_family="Courier New",
        font=dict(size=12, color="black"),
        bgcolor="LightGray",
        bordercolor="Black",
        borderwidth=1
    ),
    legend_title_text='Legend'
)

plotly.offline.iplot(fig)
```
